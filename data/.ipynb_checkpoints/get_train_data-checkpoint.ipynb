```python
import numpy as np
import matplotlib.pyplot as plt
import descarteslabs as dl
from shapely.geometry import Polygon, MultiPolygon
```

```python
geometry = Polygon([[-76.96543719924524,18.40587321886635], 
                    [-76.94758441604212,18.117661981173136], 
                    [-77.2977736250265,18.10983058728387], 
                    [-77.3094465986593,18.411085293810753], 
                    [-76.96543719924524,18.40587321886635]])
```

```python
scenes, geoctx = dl.scenes.search(
     geometry,
     products=["sentinel-2:L1C"],
     start_datetime="2014-01-01",
     end_datetime="2018-01-01",
     cloud_fraction=0.05)
```

```python
scenes
ndarray_stack = scenes.stack("red green blue",geoctx.assign())
```

```python
ndarray_stack.shape
```

```python
tile_size= 512
number_of_out_images = ((round(ndarray_stack.shape[3]/tile_size)+5)**2)
tile_stack = np.zeros([ndarray_stack.shape[0], number_of_out_images, 3, tile_size, tile_size])
count = 0
image_no = 0
for img in ndarray_stack:
    image_tiles = []
    for y in range(tile_size, img.shape[2], tile_size):
        for x in range(tile_size, img.shape[1], tile_size):
            try:
                #tile_stack[image_no, count] = img[:,x:x+tile_size, y:y+tile_size]
                image_tiles.append(img[:,x:x+tile_size, y:y+tile_size])
                count+=1
            #images at edges not correct size and are discarded
            except ValueError: 
                pass
    try:
        image_list = np.concatenate(image_tiles)    
    except ValueError:
        pass
    image_no+=1       
```
